```python
# Gradient Descent and Stochastic Gradient Descent Example Notebook

# Import Libraries
import numpy as np
import sys
import matplotlib.pyplot as plt
sys.path.append("src")
from gradient_descent import GradientDescent, StochasticGradientDescent

# Generate Synthetic Data
np.random.seed(42)
X = np.random.rand(100, 1)
y = 3 * X[:, 0] + 4 + np.random.randn(100)

# Visualize the Data
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color="blue", label="Data Points")
plt.title("Synthetic Data for Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Train Using Gradient Descent
gd_model = GradientDescent(learning_rate=0.1, epochs=1000)
gd_model.fit(X, y)
gd_predictions = gd_model.predict(X)

# Train Using Stochastic Gradient Descent
sgd_model = StochasticGradientDescent(learning_rate=0.01, epochs=50)
sgd_model.fit(X, y)
sgd_predictions = sgd_model.predict(X)

# Plot Predictions
plt.figure(figsize=(10, 8))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, gd_predictions, color="red", label="Gradient Descent Predictions")
plt.plot(X, sgd_predictions, color="green", label="Stochastic Gradient Descent Predictions")
plt.title("Model Predictions vs Actual Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Evaluation Metrics
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Print MSE Results
print(f"Gradient Descent MSE: {mean_squared_error(y, gd_predictions):.4f}")
print(f"Stochastic Gradient Descent MSE: {mean_squared_error(y, sgd_predictions):.4f}")
```
